**Q10.
A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21].
Find the mean, median, mode, range, and standard deviation. What insights can these measures provide
about the player’s scoring performance?**

**Ans:**

image.png

 **Excel Formulas Used**

Mean (Average):

=AVERAGE(A1:A8)


Median:

=MEDIAN(A1:A8)


Mode:

=IFERROR(MODE.SNGL(A1:A8),"No mode")


Range:

=MAX(A1:A8) - MIN(A1:A8)


Standard Deviation:

=STDEV.S(A1:A8)

Sorted data: 15, 17, 18, 19, 20, 21, 22, 25

**Mean:**

Add all values: 15 + 17 + 18 + 19 + 20 + 21 + 22 + 25 = 157

Divide by total games: 157 ÷ 8 = 19.625

Mean ≈ 19.63 points

 **Median :**

Middle two values (4th = 19, 5th = 20)

Median = (19 + 20) / 2 = 19.5 points

**Mode :**

All values appear only once

No mode

**Range :**

max = 25

min = 15

Range = 25 − 15 = 10 points

 **Standard Deviation:**

Measures how far scores are from the mean

Excel result: 3.113909

SD ≈ 3.11 points

This is small compared to the mean (20)

 **Insights:**

The player’s scores stay close together because the SD is small (≈ 3).

Scores do not jump wildly  they vary only within a small range (15 to 25).

Mean (19.63) and median (19.5) are almost the same, showing the center of the data is around 20 points.

No mode indicates the player does not score the same number repeatedly their scores vary slightly each game.

Range of 10 shows the maximum difference between best and worst game is moderate, not huge.

Overall, the player is consistent, usually scoring around 20 points with predictable variation.

There are no extreme outliers, so the scoring pattern is stable and reliable.
